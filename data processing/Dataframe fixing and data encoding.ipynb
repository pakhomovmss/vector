```python
from binencode import bin_encoding
import numpy as np
import pandas as pd
import time
```

```python
df = pd.read_csv('E:\\Data Science\\Datasets\\EatingData\\result_table.csv')
```

```python
# removing rows with complex id == 4, cause it's not a complex actually

df.drop(np.where(df['CmpxId'] == 4)[0], inplace=True)
df
```

```python
depths = [1, 1, 1]
depths += [9 for i in range(18)]
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
start_time = time.time()
df = bin_encoding(df, depths=depths)
print(time.time() - start_time)
```

```python
df.to_csv('E:\\Data Science\\Datasets\\EatingData\\encoded_data_0410.csv', index=False)
```
